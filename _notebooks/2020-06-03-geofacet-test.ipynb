# Panel Chart of Unites States

######                                                        

```r
#collapse-hide

######### Dependencies ##########

# install.packages("geofacet")

# https://blog.revolutionanalytics.com/2015/09/resizing-plots-in-the-r-kernel-for-jupyter-notebooks.html
library(repr)
options(repr.plot.width = 14)

library(ggplot2)
library(geofacet)

######### Plot ##########

ggplot(state_ranks, aes(variable, rank, fill = variable)) +
  geom_col() +
  coord_flip() +
  facet_geo(~ state) +
  theme_bw()
```

# Another Panel Chart of Unites States

######                                                              

```r
#collapse-hide

######### Plot ##########

ggplot(state_unemp, aes(year, rate)) +
  geom_line() +
  facet_geo(~ state, grid = "us_state_grid2") +
  scale_x_continuous(labels = function(x) paste0("'", substr(x, 3, 4))) +
  ylab("Unemployment Rate (%)")
```

# Panel Chart of European Union

######                                                             

```r
#collapse-hide

######### Plot ##########

ggplot(eu_gdp, aes(year, gdp_pc)) +
  geom_line(color = "steelblue") +
  facet_geo(~ name, grid = "eu_grid1", scales = "free_y") +
  scale_x_continuous(labels = function(x) paste0("'", substr(x, 3, 4))) +
  ylab("GDP Per Capita in Relation to EU Index (100)") +
  theme_bw()
```
